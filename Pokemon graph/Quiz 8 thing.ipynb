```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
pokemon = pd.read_csv("pokemon.csv")
fig, axes = plt.subplots(nrows = 1,ncols = 2,figsize=(12,6))
x1 = pokemon.iloc[:,6]
y1 = pokemon.iloc[:,7]
x2 = pokemon.iloc[:,]
axes[0].scatter([x1],[y1])
axes[0].set_xlabel('Attack')
axes[0].set_ylabel('Defense')
axes[0].set_title('Attack vs Defense for Pokemon')
axes[1].plot()
axes[1].set_title('Generations of Pokemon')
current_palette = sb.color_palette()
sb.countplot(x = 'Generation',data = pokemon,palette =current_palette,)

plt.show()
```
