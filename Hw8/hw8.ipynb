```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
def sinewave(x,w,a):
    ylist = []
    for i in range(len(x)):
        z = (a*mt.sin(x[i]*w))
        ylist.append(z)
    return ylist
x = np.arange(0,10,0.01)
y = sinewave(x,1,1)
print(x)

plt.figure()
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
x = np.arange(0,2*(np.pi),(2*(np.pi))/1000)
a = [0.5, 1, 1.5, 2, 2.5]
w = [0.5, 1, 1.5, 2, 2.5]
def sinewave(x,w,a):
    ylist = []
    for i in range(len(x)):
        z = (a*mt.sin(x[i]*w))
        ylist.append(z)
    return ylist
y1 = sinewave(x,w[0],a[0])
y2 = sinewave(x,w[1],a[1])
y3 = sinewave(x,w[2],a[2])
y4 = sinewave(x,w[3],a[3])
y5 = sinewave(x,w[4],a[4])
print(x)

plt.figure()
plt.plot(x,y2,label = 'normal')
plt.plot(x,y1,label = 'not normal1')
plt.plot(x,y3,label = 'not normal 2')
plt.plot(x,y4, label = 'not normal 3')
plt.plot(x,y5,label = 'not normal 4')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Sine")
plt.show()
```

```python
import numpy as np
import pandas as pd
Fred = pd.read_csv("stars.csv")
print(Fred)
```

```python
X = np.arange(0,5)
print(Fred.iloc[X,:])

df = pd.DataFrame(Fred)
NumRows = df.shape[0]
NumColumns = df.shape[1]
print("")
print("NumRows: " + str(NumRows))
print("NumColumns: " + str(NumColumns))
```

```python
def AverageMass():
    i = 0
    MassColumn = df.iloc[:,2]
    for n in range(len(MassColumn)):
        i +=MassColumn[n]
    n=i/len(MassColumn)
    return n
AverageMass()
```

```python
def AverageTemp():
    i=0
    RadiusColumn = df.iloc[:,4]
    for n in range(len(RadiusColumn)):
        i +=RadiusColumn[n]
    n = i/len(RadiusColumn)
    return n
AverageTemp()
```
